```python
import numpy as np
import matplotlib.pyplot as plt
import os
plt.style.use('./src/graph_preset.mplstyle')

'''
 * Author: Ryosuke AKIZAWA (ryosuke.akizawa@ipmu.jp)
 * Original Author: Takumi IZAWA (takumi.izawa@ipmu.jp)
 * Date: 2025-02-07
 * Version: 1.0
 * Description: for Kashiwa-Science-Camp
 * Internal Use Only
'''
```

```python
filenames = os.listdir("./data/20250214/")
for i in range (len(filenames)):
    print(f"[{i}]", filenames[i])
```

```python
file = filenames[3] # ここをかえる

npz_data =  np.load('./data/20250214/'+file,)
for key in npz_data.files:
    print(f"  {key:<8}: (rows, columns) = {npz_data[key].shape}")

data = dict(npz_data)
#data['data'] = 10**(data['data']/10)
```

オレンジ

```python
fig,ax = plt.subplots(figsize=(8,6),dpi=60)
ax.scatter(data['freq'],data['trans'],label='s-pol',color='r',s=8)
#ax.set_ylim(-0.05,1.1)
ax.set_xlim(90,140)
ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('Transmittance')
#ax.set_title(file)
ax.plot([],[],label="Apple", color="none")
ax.legend(loc='upper right')
plt.savefig("./output/"+file+".png")
```

```python
fig,ax = plt.subplots(figsize=(8,6),dpi=60)
ax.scatter(data['freq'],np.unwrap(np.rad2deg(data['phase']))-np.rad2deg(data['phase'])[0],label='s-pol',color='r',s=8)
#ax.set_ylim(-0.05,1.1)
ax.set_xlim(90,140)
ax.set_xlabel('Frequency [GHz]')
ax.set_ylabel('Relative Phase [deg]')
#ax.set_title(file)
#ax.plot([],[],label=file, color="none")
ax.legend(loc='lower right')
plt.savefig("./output/"+file+".png")
```
